```python
# Title: EDA Exploration and importing Libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Data Loading
df = pd.read_csv('../Data/Titanic-Dataset.csv')
df.head()
```

```python
# Basic Information about the dataset and null values
df.info()
df.describe(include='all').T
df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False)
```

```python
# Summary Statistics
df.describe()
df.describe().to_csv('../outputs/summary_statistics.csv')
```

```python
# Destribution plots and visualizations can be added here
plt.figure()
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
```

```python
# Boxplot outlier detection
plt.figure()
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
```

```python
# Pairplot and Correlation Heatmap
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'SibSp']].dropna())
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df[['Age', 'Fare', 'Pclass', 'SibSp', 'Survived']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Patterns and insights can be further explored here
print(df.groupby('Pclass')['Survived'].mean())
print(df.groupby('Sex')['Survived'].mean())
```

```python
# Saved Summary Statistics to CSV
df.describe().to_csv('../outputs/summary_statistics.csv')
df.to_csv('../outputs/cleaned_eda_titanic_data.csv', index=False)
```
